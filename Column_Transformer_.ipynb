# Train Test Split

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(df.drop(columns=['has_covid']),df['has_covid'],test_size=0.2)
```

# **Use Column Transformer**

```python
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OrdinalEncoder, OneHotEncoder
trf = ColumnTransformer(
    transformers=[
        ('trf1', SimpleImputer(),['fever']),
        ('trf2', OrdinalEncoder(categories = [['Mild','Strong']]),['cough']),
        ('trf3', OneHotEncoder(drop = 'first', sparse_output = False),['gender','city']),

    ],
    remainder = 'passthrough')
```

```python
trf
```

```python
X_train_trf = trf.fit_transform(X_train)
X_test_trf = trf.transform(X_test)
```

```python
X_train_trf
```

```python
X_train_trf = pd.DataFrame(data=X_train_trf, columns=trf.get_feature_names_out())
X_test_trf = pd.DataFrame(data=X_test_trf, columns=trf.get_feature_names_out())
```
